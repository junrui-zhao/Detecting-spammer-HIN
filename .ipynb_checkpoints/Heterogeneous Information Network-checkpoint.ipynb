```python
def pp1_res(text):
    count_p = text.count(u"我")
    count_e = float(text.count(u"！"))/float(len(text))
    return count_p,count_e
```

```python
def nonzero_count(mat):
    bbb = Counter(np.nonzero(mat)[0])
    i = 0
    aaa = []
    for i in range(len(mat)):
        if bbb.has_key(i):
            aaa.append(bbb[i])
        else:
            aaa.append(0)
    return aaa
```

```python
#ACS,MCS,PP1,RES
def pp1_res(text):
    count_p = text.count(u"我")/float(len(text))
    count_e = float(text.count(u"！"))/float(len(text))
    return count_p,count_e

for key,value in zip(text_id,review_text):
    sum_sm = 0
    avg_sm = 0
    max_sm = 0
    #writer_2.writerow([key])
    for i in range(0,len(value)):
        for j in range(i+1,len(value)):
            #writer_2.writerow([i])
            #writer_2.writerow([j])
            #if type(value[i]) == type(0.5) or type(value[j]) == type(0.5):
            #    writer_3.writerow([type(value[i]),type(value[j])])
            #    writer_3.writerow([value[i],value[j]])
            sm1 = Levenshtein.jaro(value[i],value[j])
            #writer_2.writerow([sm1])
            sum_sm += sm1
            if sm1>max_sm:
                max_sm = sm1
    if sum_sm !=0:
        avg_sm = float(sum_sm) / float(len(value)*(len(value)-1)/2)  
    for row in test_list:
        if row[0] == key: 
            row.append(math.floor(avg_sm*20)/20.0)
            row.append(math.floor(sum_sm*20)/20.0)
            row.append(math.floor(pp1_res(row[2])[0]*20)/20.0)
            row.append(math.floor(pp1_res(row[2])[1]*20)/20.0)
            #if avg_sm!=0:
            #    print row[0],row[1],row[2],row[6],row[7:]
            #writer_1.writerow([row])
```

```python
for value in review_time:
    value.sort()

#计算BST、ETF
for key,value in zip(time_id,review_time):
    first = value[0]
    last = value[-1]
    fl_interval = 0
    if len(value)>1 and (last-first).days<28:
        fl_interval = 1-float((last-first).days)/28.0    

    for j in range(0,len(value)):
        interval = 0
        now = value[j]    
        if len(value)>1 and now!=first and (now-first).days<7:
            interval = 1-float((now-first).days)/7.0
        for row in test_list:
            if row[0] == key and row[3] == datetime.datetime.strftime(value[j],'%Y-%m-%d %H:%M:%S'):
                row.append(math.floor(fl_interval*20)/20.0)
                row.append(math.floor(interval*20)/20.0)
               
```

```python
#计算评论同一商品的概率RSP
for key,value in zip(prod_id,review_prod):
    c = Counter(value)
    for row in test_list:
        if row[0] == key:
            if len(value) == 1:
                row.append(0)
            else :
                c.most_common(1)[0][1]
                same_product = float(c.most_common(1)[0][1])/float(len(value)+1)
                row.append(math.floor(same_product*20)/20.0)
```
