```python
#计算acs、mcs
for key,value in t.items():
    max_sm = 0
    sum_sm = 0
    avg_sm = 0
    writer_2.writerow([key])
    for i in range(0,len(value)):
        for j in range(i+1,len(value)):
            writer_2.writerow([i])
            writer_2.writerow([j])
            #if type(value[i]) == type(0.5) or type(value[j]) == type(0.5):
            #    writer_3.writerow([type(value[i]),type(value[j])])
            #    writer_3.writerow([value[i],value[j]])
            sm1 = Levenshtein.jaro(value[i],value[j])
            writer_2.writerow([sm1])
            sum_sm += sm1
            if sm1>max_sm:
                max_sm = sm1
    if sum_sm !=0:
        avg_sm = float(sum_sm) / float(len(value)*(len(value)-1)/2)  
    for row in test_list:
        if row[0] == key: 
            row.append(avg_sm)
            row.append(sum_sm)
    # print key,"avg:",avg_sm,",max:",max_sm
```

```python
#计算第一人称数量,感叹号比例
for key,value in t.items():
    for text in value:
        count_p = text.count(u"我")
        count_e = float(text.count(u"！"))/float(len(text))
        for row in test_list:
            if row[0] == key and row[2] == text:
                row.append(count_p)
                row.append(count_e)
```

```python
#评论时间
p = {}
for review in test_list:
    if p.has_key(review[0]):
        p.get(review[0]).append(datetime.datetime.strptime(review[3], "%Y-%m-%d %H:%M:%S"))
    else:
        p[review[0]] = [datetime.datetime.strptime(review[3], "%Y-%m-%d %H:%M:%S")]
```

```python
for value in p.values():
    value.sort()
```

```python
#计算bst、etf
for key,value in p.items():
    f = value[0]
    l = value[-1]
    fl_interval = 0
    if (l-f).days<28:
        fl_interval = 1-float((l-f).days)/28.0
    for row in test_list:
        if row[0] == key:
            row.append(fl_interval) 
```

```python
for key,value in p.items(): 
    first = value[0]
    for j in range(0,len(value)):
        interval = 0
        now = value[j]       
        if (now-first).days<7:
            interval = 1-float((now-first).days)/7.0
        for row in test_list:
            if row[0] == key and row[3] == datetime.datetime.strftime(value[j],'%Y-%m-%d %H:%M:%S'):
                row.append(interval)
```

```python
#评论商品
g = {}
for review in test_list:
    if g.has_key(review[0]):
        g.get(review[0]).append(review[6])
    else:
        g[review[0]] = [review[6]]
```

```python
#计算评论同一商品的概率
for key,value in g.items():
    c = Counter(value)
    for row in test_list:
        if row[0] == key:
            if len(value) == 1:
                row.append(0)
            else :
                c.most_common(1)[0][1]
                row.append(float(c.most_common(1)[0][1])/float(len(value)+1))
```
